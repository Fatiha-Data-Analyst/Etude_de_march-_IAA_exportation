### **1.2.1.3 Cleaning**

#### *1.2.1.3.1 Ajout variable croissance démographique*

```python
population.head(1)
```

```python
# Calcul variable croissance démographique en % entre 2000 et 2017
population_C=population.pivot_table(index = 'Zone', columns = 'Année', values = 'Valeur')
population_C['Croissance démographique (%)']=((population_C[2017]-population_C[2000]) /(population_C[2000])*100).round(2)
population_C.head()
```

```python
# On selectionne la colonne croissance démographique 
populationC=population_C.loc[:,['Croissance démographique (%)']]
populationC.head()
```

```python
popC=population.merge(populationC, how='inner', on='Zone')
popC
```

```python
#selection de l'année 2017
pop2017=popC[(popC['Code année'] == 2017)]
pop2017.head()
```

```python
pop2017['Code année'].unique()
```

```python
#on selectionne les colonnes qui nous interesse 
cols=['Zone','Valeur','Croissance démographique (%)']
```

```python
pop = pop2017[cols]
pop.head()
```

```python
#renommer la colonne valeur en nbr habitant
pop.rename(columns={'Valeur': 'nb_habitant (Millions)'}, inplace= True)
```

```python
#Uniformiser les unités
#diviser par 1000 pour adapter l'unité et la passer en millions d'habitant
pop['nb_habitant (Millions)']=pop['nb_habitant (Millions)']/1000
pop['nb_habitant (Millions)'] = pop['nb_habitant (Millions)'].round(2)
```

####  1.2.2.3.2.Volet Commerce International

```python
#Déterminer la quantité total de volaille importer
#on filtre le fichier sur la viande de volaille
Filtre_Volaille=Pivot_DispoA.loc[Pivot_DispoA['Produit']=='Viande de Volailles']
Filtre_Volaille.head()
```

Calcul du Taux de dépendance à l'importation de la volaille
* DISPONIBILITE INTERIEUR = Production + importations - exportations + variations des stocks (baisse ou augmentation
* Taux de dépendance aux importations (TDI) = (Importation ÷ Disponibilité intérieure) x 100

```python
Filtre_Volaille['TDI volaille (%)']=round(Filtre_Volaille['Importations - Quantité']/Filtre_Volaille['Disponibilité intérieure']*100,2)
Filtre_Volaille.head(1)
```

```python
#on renomme la colonne 
Filtre_Volaille.rename(columns={'Importations - Quantité': 'Import_Volaille (quantité)','Disponibilité alimentaire en quantité (kg/personne/an)':'Disponibilité alimentaire volaille en quantité (kg/personne/an)'}, inplace= True)
Filtre_Volaille.head(1)
```

```python
Filtre_Volaille.shape
```

```python
#on supprime les colonnes origine et produit
#Suppression des colonnes inutiles pour l'analyse
Filtre_Volaille.drop(['Origine','Produit','Disponibilité de protéines en quantité (g/personne/jour)','Disponibilité intérieure'], axis=1, inplace=True)
```

```python
Filtre_Volaille.head(1)
```

**Calcul du taux d'autosuffisance alimentaire en volaille**

formule : Sur la base des chiffres de production (source wiki)

Production alimentaire domestique ÷ (Production alimentaire domestique + importations ー exportations) ×100

celà revient à diviser la production par la production intérieure

```python
Pivot_DispoV= DispoA.pivot_table(index=['Zone', 'Origine','Produit'], 
                                            columns='Élément', values='Valeur', aggfunc=sum).reset_index()
Pivot_DispoV.head()
```

```python
Filtre_V=Pivot_DispoV.loc[Pivot_DispoV['Produit']=='Viande de Volailles']
Filtre_V.head()
```

```python
#Suppression des colonnes inutiles pour l'analyse
Filtre_V.drop(['Alimentation pour touristes','Aliments pour animaux', 'Autres utilisations (non alimentaire)','Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité alimentaire en quantité (kg/personne/an)', 'Disponibilité de matière grasse en quantité (g/personne/jour)',
                'Nourriture','Pertes','Résidus','Semences', 'Traitement', 'Variation de stock','Disponibilité de protéines en quantité (g/personne/jour)','Importations - Quantité','Exportations - Quantité'], axis=1, inplace=True)
Filtre_V.head()
```

```python
Filtre_V['TAS volaille (%)']=round(Filtre_V['Production']/Filtre_V['Disponibilité intérieure']*100,2)
Filtre_V.head(5)
```

```python
Filtre_V['TAS volaille (%)'] = Filtre_V['TAS volaille (%)'].replace(np.nan, 0)
Filtre_V.head()
```

```python
#Suppression des colonnes inutiles pour l'analyse
Filtre_V.drop(['Disponibilité intérieure','Production','Produit','Origine'], axis=1, inplace=True)
Filtre_V.head()
```

####  1.2.2.3.3.Volet Régime alimentaire 

* calcul des variables
    * %Disponibilité alimentaire origine animale
    * %Protéines origine animale


* On créé un autre DF axé sur l'origine animale pour calculer le total des disponibilités et proteines d'origine animale

```python
#on filtre le fichier sur l'origine animale
Filtre_animale=Pivot_DispoA.loc[Pivot_DispoA['Origine']=='animal']
Filtre_animale.head()
```

```python
#Onregroupe le df par pays
Filtre_animal_Zone=Filtre_animale.groupby(['Zone'], as_index=False).sum()
Filtre_animal_Zone.head()
```

```python
#on supprime la colonne disponibilité intérieure et importation qui n'ont ici plus d'utilité
#Suppression des colonnes inutiles pour l'analyse
Filtre_animal_Zone.drop(['Disponibilité intérieure','Importations - Quantité'], axis=1, inplace=True)
Filtre_animal_Zone.head()
```

```python
#on renomme la colonne 
Filtre_animal_Zone.rename(columns={'Disponibilité alimentaire en quantité (kg/personne/an)': 'Disponibilité alimentaire animale totale en quantité (kg/personne/an)',"Disponibilité de protéines en quantité (g/personne/jour)":"Disponibilité de protéines animale totale en quantité (g/personne/jour)"}, inplace= True)
Filtre_animal_Zone.head(1)
```

```python
#on créé un autre df qui permette de calculer le total des dispo par pays toute origine confondu
dispototal=Pivot_DispoA
dispototal.head(1)
```

```python
dispototal.drop(['Origine','Produit','Disponibilité intérieure','Importations - Quantité'], axis=1, inplace=True)
dispototal.head()
```

```python
#on regroupe par pays
dispototal_Zone=dispototal.groupby(['Zone'], as_index=False).sum()
dispototal_Zone.head()
```

```python
dispototal_Zone.rename(columns={'Disponibilité alimentaire en quantité (kg/personne/an)': 'Disponibilité alimentaire totale en quantité (kg/personne/an)',"Disponibilité de protéines en quantité (g/personne/jour)":"Disponibilité de protéines totale en quantité (g/personne/jour)"}, inplace= True)
dispototal_Zone.head(5)
```

```python
#jointure outer entre les fichiers 
from functools import reduce
# On réunit les DataFrames dans une liste
listedispo = [Filtre_Volaille,dispototal_Zone,Filtre_animal_Zone,Filtre_V]
# On les fusionne
Dispo= reduce(lambda  left,right: pd.merge(left,right,on=['Zone'],
                                            how='outer'), listedispo)
```

```python
Dispo.head()
```

```python
#inserons les colonnes %Disponibilité alimentaire en quantité origine animale,%Disponibilité alimentaire en quantité origine volaille,%Protéines origine animale

Dispo["Part de la dispo alimentaire d’origine animale (%)"]=round(Dispo['Disponibilité alimentaire animale totale en quantité (kg/personne/an)']/Dispo['Disponibilité alimentaire totale en quantité (kg/personne/an)']*100,2)
Dispo["Part de la dispo alimentaire d’origine volaille (%)"]=round(Dispo['Disponibilité alimentaire volaille en quantité (kg/personne/an)']/Dispo['Disponibilité alimentaire totale en quantité (kg/personne/an)']*100,2)
Dispo["Part des protéines d'origine animale (%)"]=round(Dispo['Disponibilité de protéines animale totale en quantité (g/personne/jour)']/Dispo['Disponibilité de protéines totale en quantité (g/personne/jour)']*100,2)
Dispo.head()
```

```python
#on supprime les colonnes dont nous n'avons plus besoin
Dispo.drop(['Disponibilité alimentaire volaille en quantité (kg/personne/an)','Disponibilité alimentaire totale en quantité (kg/personne/an)','Disponibilité de protéines totale en quantité (g/personne/jour)','Disponibilité alimentaire animale totale en quantité (kg/personne/an)','Disponibilité de protéines animale totale en quantité (g/personne/jour)'], axis=1, inplace=True)
Dispo.head()
```
